## Реалізація двовимірного дискретного косинусного перетворення

Матриця перетворення для випадку N = 8 :

$$ 
G_u = \sqrt{\frac{2}{N}} \frac{1}{\sqrt{2}} f_0 + 
\sqrt{\frac{2}{N}} \sum_{x=1}^7 f_x \, cos\left( \frac{\pi}{8} (u + \frac{1}{2}) x \right)
$$



Перетворення застосовується наступним чином (ДКП застосовується до кожного стовпця, а потім до кожного рядка):


$$ G = {DCT} \cdot f \cdot {DCT}^{T} $$
