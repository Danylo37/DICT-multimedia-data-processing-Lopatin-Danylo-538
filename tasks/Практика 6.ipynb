# Матриця перетворення

```python
#from IPython.display import Image
#Image(filename='dct_matrix.png')
# Ортогональна матриця має такий вигляд:
```

$$ 
С_{ij} = \sqrt{\frac{2}{N}} a_i cos\left( \frac{i(2j + 1)\pi}{2n} \right),
$$

где $i,j = 0,..,N-1$,

$$
a_i =
  \begin{cases}
    1/\sqrt2       & \quad \text{если } i=0,\\
    1  & \quad \text{если } i = 1,..,N-1
  \end{cases}
\
$$

## Одновимірне дискретне косинусне перетворення (ДКП)
для вихідного сигналу (послідовності) у вигляді вектора $x = [x_0,..,x_{n-1}]^T$ також є вектором $y = [y_0,..,y_{n-1}]^T$, де

$$
y = Cx.
$$

## Зворотне ДКП
при використанні ортогональної матриці перетворення має такий вигляд:

$$
x = C^Ty.
$$
